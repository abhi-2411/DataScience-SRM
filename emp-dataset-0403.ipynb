```python
import pandas as pd
import numpy as np

# Sample employee dataset
data = {
    "emp_id": [101, 102, 103, 104, 105],
    "name": ["Amit", "Sara", "John", "Priya", "Rahul"],
    "designation": ["Manager", None, "Developer", "Analyst", None],
    "years of experience": [5, 2, 8, 3, 6],
    "age": [30, 25, 40, 28, 35],
    "salary": [900, 1200, 1500, 800, 1100]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)
```

```python
x = 2
df["age"] = df["age"] + x
```

```python
df.rename(columns={"years of experience": "exp"}, inplace=True)
```

```python
df = df[df["salary"] <= 1000]
```

```python
df.dropna(subset=["designation"], inplace=True)
```
